### Check mask vs forcing

```python
# read solid file mask pfb
pfsol_mask = pf.read_pfb("CONUS2_scaling_48_36.out.mask.pfb")

# check dimensions and type
print(pfsol_mask.shape)
print(type(pfsol_mask))
```

```python
# flip array vertically
pfsol_mask = pfsol_mask[:, ::-1, :] # Flip array vertically
# make new array with 1s and 0s for active cells in pfsol mask 
mask_array_pfsol = np.where(pfsol_mask[0] <= 0, 0, 1)
```

### Make Forcing Tifs

```python
# this file is needed for the geotransform and projection below
file = '../Check_EvapTrans/1km_CONUS2_landcover_IGBP.tif'

# open tiff mask
lc_tif = gdal.Open(file)
type(lc_tif)
```

```python
hour = 1
ii = 0 #[Temp,DLWR,DSWR,Press,SPFH,APCP,UGRD,VGRD]
temp = pf.read_pfb(f'{jan_path}{jan1_forcing[ii]}')
force_pfb0 = force_pfb[hour, :, :]
force_pfb0[force_pfb0<0]=np.nan
force_pfb0 = np.flipud(force_pfb0)
```

```python
dst_filename = f'Hour{hour}_{names[ii]}.tiff'
x_pixels = 4442  # number of grid cells in x
y_pixels = 3256  # number of grid cells in y
driver = gdal.GetDriverByName('GTiff')
dataset = driver.Create(dst_filename,x_pixels, y_pixels, 1,gdal.GDT_Float32)
dataset.GetRasterBand(1).WriteArray(force_pfb0) ### create tif with array you want to use

# Add GeoTranform and Projection information to newly created tif file
geotrans=lc_tif.GetGeoTransform()  #get GeoTranform from existing tif file (here 'conus2tiff')
proj=lc_tif.GetProjection() #you can import or get from a existing tif   
dataset.SetGeoTransform(geotrans)
dataset.SetProjection(proj)
dataset.FlushCache()
dataset=None
```
